```python
x=[1,3,5,7,9]
y=[2,4,6,8,10]
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['age'], y=df['chol'])
plt.title('Scatter Plot')
plt.show()
```

```python
data = pd.DataFrame({
    'Math': [85, 78, 92, 88, 76],
    'Science': [88, 79, 95, 84, 82],
    'English': [75, 83, 89, 92, 80]
})

# correlation matrix
correlation_matrix = data.corr()

# heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x=df['age'],y= df['chol'])

plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Line Plot of Age vs. Cholesterol")
plt.show()
```

```python


plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['age'], y=df['chol'])
plt.title('Scatter Plot of Age vs. Cholesterol')
plt.show()
```

```python

cp_counts = df['cp'].value_counts()
cp_counts.plot(kind='bar')

plt.xlabel("Chest Pain Type")
plt.ylabel("Frequency")
plt.title("Bar Plot of Chest Pain Types")
plt.show()
```

```python

plt.hist(df['age'], bins=10)

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")
plt.show()
```

```python

target_counts = df['target'].value_counts()
target_counts.plot(kind='pie', autopct='%1.1f%%')

plt.title("Pie Chart of Heart Disease Presence")
plt.ylabel('') # To remove the default ylabel
plt.show()
```
